# Aula: Introdução ao Matplotlib e Gráficos Básicos

## 1. O que é o Matplotlib?
**Matplotlib** é uma biblioteca de visualização de dados em Python, usada para criar gráficos estáticos e interativos.

```
!pip install matplotlib
```

```
import matplotlib.pyplot as plt
```

## 2. Criando o Primeiro Gráfico (Linha)

```

# Dados de exemplo
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 19]

# Criando o gráfico de linha
plt.plot(x, y)

# Adicionando título e rótulos
plt.title('Gráfico de Linha - Exemplo')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Exibindo o gráfico
plt.show()
```

## 3. Gráficos de Barras

```

# Dados de exemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [3, 7, 2, 5, 6]

# Criando gráfico de barras
plt.bar(categorias, valores)

# Adicionando título e rótulos
plt.title('Gráfico de Barras - Exemplo')
plt.xlabel('Categorias')
plt.ylabel('Valores')

# Exibindo o gráfico
plt.show()
```

## 4. Gráficos de Dispersão (Scatter)

```

import numpy as np

# Gerando dados aleatórios
x = np.random.rand(50)
y = np.random.rand(50)

# Criando gráfico de dispersão
plt.scatter(x, y)

# Adicionando título e rótulos
plt.title('Gráfico de Dispersão')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')

# Exibindo o gráfico
plt.show()
```

## 5. Trabalhando com DataFrames (Pandas)

```

import pandas as pd

# Criando um DataFrame
dados = {
    'Mes': ['Janeiro', 'Fevereiro', 'Março', 'Abril'],
    'Vendas': [250, 300, 280, 310]
}

df = pd.DataFrame(dados)

# Criando gráfico de linha a partir do DataFrame
plt.plot(df['Mes'], df['Vendas'])

# Adicionando título e rótulos
plt.title('Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Vendas')

# Exibindo o gráfico
plt.show()
```

# Aula 2: Gráficos Avançados e Customização

## 1. Gráficos de Pizza (Pie Chart)

```

# Dados de exemplo
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Criando gráfico de pizza
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Exibindo o gráfico
plt.show()
```

## 2. Histogramas

```

# Gerando dados aleatórios
dados = np.random.randn(1000)

# Criando o histograma
plt.hist(dados, bins=30)

# Adicionando título e rótulos
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frequência')

# Exibindo o gráfico
plt.show()
```

## 3. Customizando Gráficos

```

# Exemplo de customização
x = [1, 2, 3, 4, 5]
y = [10, 12, 15, 13, 16]

plt.plot(x, y, color='green', linestyle='--', marker='o', label='Linha Exemplo')

# Adicionando legenda
plt.legend()

# Salvando o gráfico
plt.savefig('grafico_customizado.png')

# Exibindo o gráfico
plt.show()
```
