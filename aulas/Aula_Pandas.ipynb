# **Aula: Introdução ao Pandas para Ciência de Dados**

**1. Introdução ao Pandas**

Objetivo: Entender o que é o pandas e sua importância na ciência de dados.

**O que é o pandas?**

Pandas é uma biblioteca Python usada para manipulação e análise de dados. É uma ferramenta poderosa para lidar com dados estruturados (tabelas) e não estruturados.
Facilita a leitura, manipulação e limpeza de dados, essencial para qualquer projeto de ciência de dados.

**Estruturas de Dados Principais:**

**Series:** É uma estrutura unidimensional que pode ser vista como uma coluna de uma tabela. Cada elemento da Series tem um índice.

**DataFrame:** É uma estrutura bidimensional, como uma tabela de dados com linhas e colunas. É a estrutura mais utilizada no pandas.

**2. Criação de DataFrames e Series**

Objetivo: Aprender a criar e manipular Series e DataFrames.

**Criando uma Series:**

Exemplo 1: Criando uma Series a partir de uma lista.

```python
import pandas as pd

lista = [1, 2, 3, 4, 5]
serie = pd.Series(lista)
print(serie)
```

Exemplo 2: Criando uma Series a partir de um dicionário.

```python
dados = {'a': 1, 'b': 2, 'c': 3}
serie = pd.Series(dados)
print(serie)
```

**Criando um DataFrame:**

Exemplo 1: Criando um DataFrame a partir de um dicionário de listas.

```python
dados = {
    'Nome': ['Ana', 'João', 'Maria'],
    'Idade': [23, 45, 35]
}
df = pd.DataFrame(dados)
print(df)
```

Exemplo 2: Criando um DataFrame a partir de listas de dicionários.

```python
dados = [
    {'Nome': 'Ana', 'Idade': 23},
    {'Nome': 'João', 'Idade': 45},
    {'Nome': 'Maria', 'Idade': 35}
]
df = pd.DataFrame(dados)
print(df)
```

**4. Manipulação de Dados**

Objetivo: Aprender a selecionar, filtrar, e transformar dados em um DataFrame.

**Selecionando e Filtrando Dados:**

Exemplo 1: Selecionando uma coluna.

```python
nomes = df['Nome']
print(nomes)
```

Exemplo 2: Filtrando linhas com base em uma condição.

```python
maiores_30 = df[df['Idade'] > 30]
print(maiores_30)
```

**Alterando Estruturas:**

Exemplo 1: Adicionar uma nova coluna.

```python
df['Cidade'] = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']
print(df)
```

Exemplo 2: Remover uma coluna.

```python
df = df.drop('Cidade', axis=1) print(df)
```

Exemplo 3: Renomeando colunas.

```python
df = df.rename(columns={'Nome': 'Nome Completo', 'Idade': 'Anos'})
print(df)
```

**Tratamento de Dados Faltantes:**

Exemplo 1: Detectando dados faltantes.


```python
print(df.isnull())
```

Exemplo 2: Removendo dados faltantes.

```python
df = df.dropna()
print(df)
```

Exemplo 3: Preenchendo dados faltantes.

```python
df = df.fillna(0)
print(df)
```

**5. Operações em DataFrames**

Objetivo: Executar operações matemáticas, agrupamentos e junções em DataFrames.

**Operações Básicas:**

Exemplo 1: Calculando a média de uma coluna.

```python
media_idade = df['Idade'].mean()
print(f'Média de idade: {media_idade}')
```

Exemplo 2: Aplicando uma função a uma coluna.

```python
df['Idade em 10 Anos'] = df['Idade'].apply(lambda x: x + 10)
print(df)
```

**Agrupamento e Agregação:**

Exemplo 1: Agrupando dados por uma coluna.

```python
grupo = df.groupby('Cidade')['Idade'].mean()
print(grupo)
```

Exemplo 2: Agregação múltipla.

```python
agregados = df.groupby('Cidade').agg({'Idade': ['mean', 'sum']})
print(agregados)
```

**Mesclagem e Junção de Dados:**

Exemplo 1: Concatenando dois DataFrames.

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df_concatenado = pd.concat([df1, df2])
print(df_concatenado)
```

Exemplo 2: Mesclando/juntando dois DataFrames.

```python
df1 = pd.DataFrame({'Nome': ['Ana', 'João'], 'Idade': [23, 45]})
df2 = pd.DataFrame({'Nome': ['Ana', 'João'], 'Cidade': ['SP', 'RJ']})
df_merge = pd.merge(df1, df2, on='Nome')
print(df_merge)
```

**6. Trabalhando com Dados Temporais**

Objetivo: Manipular e analisar dados temporais.

**Data e Hora:**

Exemplo 1: Convertendo uma coluna para datetime.

```python
df['Data'] = pd.to_datetime(df['Data'])
print(df)
```

Exemplo 2: Extraindo componentes de data.

```python
df['Ano'] = df['Data'].dt.year
df['Mes'] = df['Data'].dt.month
print(df)
```

**Resampling e Operações com Frequências Temporais:**

Exemplo 1: Mudança de frequência temporal.

```python
df_resampled = df.set_index('Data').resample('M').sum()
print(df_resampled)
```

Exemplo 2: Operações com rolling e shifting.

```python
df['Média Móvel'] = df['Valores'].rolling(window=3).mean()
print(df)
```
